```python
import pandas as pd
import numpy as np
import plotly.express as px
data = pd.read_csv("/content/drive/MyDrive/Data science/online transaction fraud detection.csv")
data.head(10)
```

```python
data.isnull().sum()
```

```python
data.type.value_counts()
```

```python
type = data["type"].value_counts()
transactions = type.index
quantity = type.values
figure = px.pie(data,
             values=quantity,
             names=transactions,hole = 0.1,
             title="Distribution of Transaction Type")
figure.show()
```

```python
# Checking correlation
correlation = data.corr()
correlation["isFraud"].sort_values(ascending=False)
```

```python
data["type"] = data["type"].map({"CASH_OUT": 1, "PAYMENT": 2,
                                 "CASH_IN": 3, "TRANSFER": 4,
                                 "DEBIT": 5})
data["isFraud"] = data["isFraud"].map({0: "No Fraud", 1: "Fraud"})
data.head()
```

```python
# splitting the data
from sklearn.model_selection import train_test_split
x = np.array(data[["type", "amount", "oldbalanceOrg", "newbalanceOrig"]])
y = np.array(data[["isFraud"]])
```

```python
from sklearn.metrics import accuracy_score
# training a machine learning model
from sklearn.tree import DecisionTreeClassifier
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.10, random_state=42)
model = DecisionTreeClassifier()
model.fit(xtrain, ytrain)
score1 = model.score(xtest, ytest)
accuracy_score = score1*100
print("The accurancy of the model is",accuracy_score)
```

```python
# prediction
#features = [type, amount, oldbalanceOrg, newbalanceOrig]
features = np.array([[4, 9000.60, 9000.60, 0.0]])
model.predict(features)
```
